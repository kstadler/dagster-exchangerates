```python
import dagstermill

context = dagstermill.get_context(solid_config='../../exchangerates.sqlite')
db_file = context.solid_config
print(db_file)
```

```python
import sqlite3
import pandas as pd
import plotly.express as px

conn = sqlite3.connect(db_file)
df = pd.read_sql_query("SELECT day, rate FROM exchangerates WHERE currency = 'USD' ORDER BY day", conn)
conn.close()

fig = px.line(df, x="day", y="rate", title='EUR - USD Exchange Rate')
fig.show()
```
